```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
import pandas as pd
```

```python
# Import data sheet
file_data = open('orange.txt', 'r').read()

# Removing dots
file = file_data.split('.')
lists = [x.replace('\n', '') for x in file]

vector = TfidfVectorizer(stop_words='english')
res = vector.fit_transform(lists)

print(res)
```

```python
fn = vector.get_feature_names_out()
fn
```

```python
dataframe = pd.DataFrame(res.todense().T, index=fn, columns=[f'Doc{i+1}' for i in range(len(lists))])
dataframe
```
